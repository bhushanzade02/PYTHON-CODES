```python
# ### Problem 2.2: Data Type Conversion
# *Task*: Fix data types in this problematic dataset.
# 1. Convert Price to float
# 2. Convert Quantity to integer
# 3. Convert Is_Weekend to boolean
# 4. Convert Date to datetime
# sample:
sales_data = pd.DataFrame({
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03'],
    'Price': ['29.99', '15.50', '89.00'],  # Should be numeric
    'Quantity': ['5', '3', '7'],           # Should be numeric
    'Is_Weekend': ['True', 'False', 'False']  # Should be boolean
})

print(" 1. Convert Price to float","\n")
sales_data.head()


print("before conversion")
print(sales_data.dtypes,"\n")

sales_data['Price'] = sales_data['Price'].astype(float)
sales_data['Quantity'] = sales_data['Quantity'].astype(int)
sales_data['Is_Weekend'] = sales_data['Is_Weekend'].astype(bool)
sales_data['Date'] = pd.to_datetime(sales_data['Date'])

print("after conversion")
print(sales_data.dtypes,"\n")

print(sales_data)
```

```python
# ### Problem 3.1: Complex Filtering
# *Task*: Using the BMW dataset, find cars that meet specific criteria.

# *Your Tasks:*
# 1. Find all cars from 2016 or later
# 2. Find cars with price between 15000 and 25000
# 3. Find diesel cars with mileage less than 40000
# 4. Find the cheapest car for each fuel type



df = pd.read_csv(r"D:\09_DOWNLOADS (BRAVE)\bmw_with_missing_data.csv")

df['year'] = pd.to_numeric(df['year'], errors='coerce')
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['mileage'] = pd.to_numeric(df['mileage'], errors='coerce')


df = df.dropna()


cars_2016_and_later = df[df['year'] >= 2016]
print("Cars from 2016 or later:\n", cars_2016_and_later, "\n")

cars_price_range = df[(df['price'] >= 15000) & (df['price'] <= 25000)]
print("Cars with Price between 15000 and 25000:\n", cars_price_range, "\n")


diesel_low_mileage = df[(df['fuelType'] == 'Diesel') & (df['mileage'] < 40000)]
print("Diesel cars with mileage < 40000:\n", diesel_low_mileage, "\n")

```

```python
# ### Problem 3.2: Grouping and Aggregation
# *Task*: Analyze the BMW dataset by groups.

# *Your Tasks:*
# 1. Group by fuel type and find average price
# 2. Group by year and count how many cars
# 3. Group by transmission type and find max, min, and average mileage
# 4. Find the most expensive car in each model category


print("1️ Group by fuel type and find average price")
avg_price_fuel = df.groupby('fuelType')['price'].mean()
print(avg_price_fuel, "\n")


print("2️ Group by year and count how many cars")
cars_per_year = df.groupby(['year','price'])['model'].count()
print(cars_per_year, "\n")


print("3️ Group by transmission type and find max, min, and average mileage\n")
mileage_stats = df.groupby('transmission')['mileage'].agg(['max', 'min', 'mean'])
print(mileage_stats, "\n")


print("4️ Find the most expensive car in each model category\n")
most_expensive_per_model = df.loc[df.groupby('model')['price'].idxmax(), ['model', 'price', 'fuelType', 'year']]
print(most_expensive_per_model)
```
