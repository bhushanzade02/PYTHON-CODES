```python
df4 = pd.DataFrame({"id":["1","2","3","4","5","6","7"], "num":["1","1","1","2","1","2","2"]})
```

```python
df4
```

```python
#look at the no before and after each row 
df4['prev'] = df4['num'].shift(1) 
df4['next'] = df4['num'].shift(-1)



same_three = df4[(df4['num'] == df4['prev'] ) & (df4['num'] == df4['next'])]

result = same_three['num'].drop_duplicates()

final_2 = pd.DataFrame({'':result})

final_2
```
